```python
ace = 1
jack = 10
queen = 10
king = 10

possible_cards = [ace, 2, 3, 4, 5, 6, 7, 8, 9, 10, jack, queen, king]
```

Create shoe out of possible cards, shuffle and place cut card 0.75 to 1.5 decks (remove all cards after cut card)

```python
n_decks = 6
deck_len = 52
lower_cut_bound = 0.75
upper_cut_bound = 1.5

shoe = possible_cards * 4 * n_decks
random.shuffle(shoe)
print(len(shoe))
print("Top Card: " + str(shoe[0]))

cut = random.randint(int(lower_cut_bound * deck_len), int(upper_cut_bound * deck_len))
print("Cut: " + str(cut))
shoe = shoe[:-cut]
print(len(shoe))
print("Top Card: " + str(shoe[0]))
```
